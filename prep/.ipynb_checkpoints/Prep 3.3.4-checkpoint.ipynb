```python
import numpy as np
import matplotlib.pyplot as plt

poisson = np.random.poisson(5, 100)
plt.hist(poisson)
```

```python
plt.hist(poisson)
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(poisson.mean() - poisson.std(), color='b', linestyle='dashed', linewidth=1)

plt.show()
```

```python
normal = np.random.normal(0, 100, 100)
plt.hist(normal)
```

```python
plt.hist(normal)
plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal.mean() + normal.std(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(normal.mean() - normal.std(), color='b', linestyle='dashed', linewidth=1)

plt.show()
```
